# Task 1 â€” Exploring and Visualizing the Iris Dataset

**Objective:** Load the Iris dataset, inspect its structure, visualize features (scatter, histogram, boxplot, correlation), and write short observations about what we find.


```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Make plots show in the notebook
%matplotlib inline

# Load iris dataset from seaborn
df = sns.load_dataset('iris')

# Show first rows
df.head()
```

```python
# Dataset shape and columns
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# Basic info
display(df.info())

# Numeric summary
display(df.describe())
```

```python
# Check missing values
print("Missing values per column:\n", df.isnull().sum())

# Check duplicates (optional)
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Pairplot: scatter matrix colored by species
sns.pairplot(df, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pairplot of Iris features (colored by species)', y=1.02)
plt.show()
```

```python
# Scatter plot: Petal length vs Petal width
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=60)
plt.title('Petal length vs Petal width')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.show()
```

```python
# Histograms for numeric features
df[['sepal_length','sepal_width','petal_length','petal_width']].hist(bins=12, figsize=(10,6))
plt.suptitle('Histograms of numeric features')
plt.show()
```

```python
# Scatter plot: Petal length vs Petal width
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=60)
plt.title('Petal length vs Petal width')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.show()
```

```python
# Histograms for numeric features
df[['sepal_length','sepal_width','petal_length','petal_width']].hist(bins=12, figsize=(10,6))
plt.suptitle('Histograms of numeric features')
plt.show()
```

```python
# Boxplots: feature distribution by species
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Sepal length by species')

plt.subplot(2,2,2)
sns.boxplot(x='species', y='sepal_width', data=df)
plt.title('Sepal width by species')

plt.subplot(2,2,3)
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal length by species')

plt.subplot(2,2,4)
sns.boxplot(x='species', y='petal_width', data=df)
plt.title('Petal width by species')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr = df.drop('species', axis=1).corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation')
plt.show()

corr
```

# EDA Summary & Insights

- Dataset size: 150 rows, 5 columns. No missing values.
- Petal length and petal width show clear separation between species - they are strong discriminators.
- Sepal measurements overlap more across species.
- Petal length and petal width are strongly positively correlated.
- No significant outliers that would force removal (check boxplots).


```python
# Example: save the petal scatter plot to file
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=60)
plt.title('Petal length vs Petal width')
plt.savefig('petal_scatter.png', bbox_inches='tight', dpi=150)
plt.show()
```
